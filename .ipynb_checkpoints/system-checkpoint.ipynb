```python
# systems.py

import numpy as np
from abc import ABC, abstractmethod

class GeneralSystem(ABC):
    """
    Abstract base class for a general stochastic system
    as requested in the slide.
    """
    @abstractmethod
    def dynamics(self, x, u):
        """Evolves the state one time step."""
        pass
    
    @abstractabstractmethod
    def likelihood(self, y, x):
        """Calculates the probability of measurement y given state x."""
        pass

class LTISystem(GeneralSystem):
    """
    Implements a Linear Time-Invariant (LTI) system:
    x_k+1 = A*x_k + B*u_k + w_k
    y_k   = C*x_k + v_k
    
    Slide notation:
    Q (process noise cov) = Sigma_x
    R (measurement noise cov) = Sigma_y
    """
    
    def __init__(self, A, B, C, Q, R):
        self.A = np.asarray(A)
        self.B = np.asarray(B)
        self.C = np.asarray(C)
        self.Q = np.asarray(Q) # Process noise covariance (Sigma_x)
        self.R = np.asarray(R) # Measurement noise covariance (Sigma_y)
        
        self.state_dim = self.A.shape[0]
        self.control_dim = self.B.shape[1]
        self.measurement_dim = self.C.shape[0]
        
    def dynamics(self, x, u, add_noise=True):
        """
        Calculates the next state x_k+1
        """
        x = np.asarray(x).reshape((self.state_dim, 1))
        u = np.asarray(u).reshape((self.control_dim, 1))
        
        # Deterministic part
        x_next = self.A @ x + self.B @ u
        
        # Add process noise w_k ~ N(0, Q)
        if add_noise:
            w_k = np.random.multivariate_normal(
                np.zeros(self.state_dim), self.Q
            ).reshape((self.state_dim, 1))
            x_next += w_k
            
        return x_next

    def measurement(self, x, add_noise=True):
        """
        Calculates the measurement y_k
        """
        x = np.asarray(x).reshape((self.state_dim, 1))
        
        # Deterministic part
        y = self.C @ x
        
        # Add measurement noise v_k ~ N(0, R)
        if add_noise:
            v_k = np.random.multivariate_normal(
                np.zeros(self.measurement_dim), self.R
            ).reshape((self.measurement_dim, 1))
            y += v_k
            
        return y

    def likelihood(self, y, x):
        """
        Calculates p(y | x). Not needed for simulation,
        but included as requested.
        """
        y_pred = self.C @ x
        error = y - y_pred
        inv_R = np.linalg.inv(self.R)
        det_R = np.linalg.det(self.R)
        
        norm_const = 1.0 / np.sqrt(((2 * np.pi) ** self.measurement_dim) * det_R)
        exponent = -0.5 * error.T @ inv_R @ error
        
        return norm_const * np.exp(exponent)
```
